### Tarefa 2. Descubra qual a idade média por ocupação.

*Dica: use a função groupby do Pandas na coluna occupation*

````python
df = pd.read_csv('/content/users_dataset.csv', sep='|')
````

````python
df[['age', 'occupation']].groupby(['occupation']).mean().round(0)
````

### Tarefa 3. Para cada ocupação descubra a idade mínima e máxima.

````python
df[['age', 'occupation']].groupby(['occupation']).min().round(0)
````

````python
df[['age', 'occupation']].groupby(['occupation']).max().round(0)
````
